# Explore Categorical Variables - Species

```python
average = df.groupby(by='species').mean()
average.plot(kind='bar', subplots=True, figsize= (10,15))
```

#### Virginica has the highest petal length followed by versicolor and setosa.
#### Setosa has the highest sepal width followed by virginica and Versicolor.
