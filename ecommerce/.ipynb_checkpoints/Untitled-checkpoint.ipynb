## Step 1 : defined dependent and independent variables

```python
# dependent variables
Y = customers['Yearly Amount Spent']

# independent varsiables
X = customers[['Avg. Session Length', 'Time on App', 'Time on Website', 'Length of Membership']]
```

## Step 2 : split training and test data using train_test_split and define the percentage for train test to split in


```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=101)

#X_train - 350 
#Y_train - 350 
#X_test - 150
#Y_test - 150
```

## Step 3: Build the training model
you can use .fit method to train the model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
lm.coef_
```

## Step 4: Testing our model

```python
predictions = lm.predict(X_test)
```

```python
plt.pyplot.scatter(y_test, predictions)
plt.pyplot.xlabel('Y Test')
plt.pyplot.ylabel('Predicted Y')
```

## Evaluating our model

## There are different ways to measure the error between the predicted y-values and the actual y-values. We will calculate three kinds of errors using NumPy and Scikit-learn’s metrics:

1.Mean Absolute Error 
2.Mean Squared Error
3.Root Mean Squared Error

###### The smaller the means squared error, the closer you are to finding the line of best fit. 

```python
from sklearn import metrics
```

```python
mae = metrics.mean_absolute_error(y_test, predictions)
mse = metrics.mean_squared_error(y_test, predictions)
rmse = np.sqrt(metrics.mean_squared_error(y_test, predictions))
```

```python
print "Mean Absolute Error is {}".format(mae)
print "Mean Square Error is {}".format(mse)
print "Root Mean Sqaure Error is {}".format(rmse)
```

## Step 5: Residuals - plot the histogram of the residuals

```python
sns.distplot(y_test-predictions, bins=50, kde=True)
```

## Step 6 Making the decision

```python
coeffs = pd.DataFrame(data=lm.coef_.transpose(), index=X.columns, columns=['Coefficient'])
```
